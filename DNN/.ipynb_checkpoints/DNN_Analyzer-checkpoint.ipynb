```python
#load needed things
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline
from keras.models import Sequential, Model
from keras.optimizers import SGD, Adam
from keras.layers import Input, Activation, Dense
from keras.utils import np_utils
from keras.wrappers.scikit_learn import KerasClassifier
from keras.callbacks import EarlyStopping, ModelCheckpoint
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as pyp
import itertools
import math
import ROOT
import cPickle as pickle
from UserFunctions import *

# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
```

```python
#load events
comparison = '>='
njets = 2
min4lmass = 118
max4lmass = 130

filein = open('hzz4l_{0}jets_m4l{1}-{2}GeV.pkl'.format(njets,min4lmass,max4lmass),'r')
events = pickle.load( filein )
filein.close()
```

```python
#creates a copy from orig events
orig_events = events
events = {}
events['VBF'] = orig_events['VBF']
events['HJJ'] = orig_events['HJJ']
events['ZH'] = orig_events['ZH']
events['WH'] = orig_events['WH']
events['ttH'] = orig_events['ttH']
events['qqZZ'] = orig_events['qqZZ']
events['ggZZ'] = orig_events['ggZZ']
events['TTZ'] = orig_events['TTZ']
```

```python
#to select how many jets to use;  eg. 7 = 4leptons + 3jets
nparticles = 7
features = {
    'pt' : None,
    'eta': None,
    'phi': None
    #'e'  : None
}
nfeatures = len(features)
```

```python
#prepare set        
inputs = {}
for ik in events:
    inputs[ik] = np.zeros((len(events[ik]), nparticles*nfeatures))
    for i in range(len(events[ik])):
        for j in range(nparticles):
            for iprop in range(nfeatures):
                if(events[ik][i][j+6][iprop] != -999.):
                    inputs[ik][i][j*nfeatures+iprop] = events[ik][i][j+6][iprop]
                else:
                    break

print 'Components: ',len(inputs)  
#print inputs['VBF'][2]
#for i in range(len(inputs['ttH'])):
#    for j in range(nparticles):
#        if(math.fabs(inputs['ttH'][i][j*nfeatures+2]) > 4):
#            print inputs['ttH'][i][j*nfeatures+2]
#print events['ttH'][34]
```

```python
opt = Adam()
dnn_model = Sequential()
dnn_model.add(Dense(21, input_shape=(nparticles*nfeatures,), activation='relu'))
dnn_model.add(Dense(13, activation='relu'))
dnn_model.add(Dense(5, activation='relu'))
dnn_model.add(Dense(1, activation='sigmoid'))
    
# load best weights
dnn_model.load_weights("weights_best_vbf_all_weighted.hdf5")
# Compile model (required to make predictions)
dnn_model.compile(loss='binary_crossentropy', optimizer=opt)
```

```python
from keras import backend as K

# with a Sequential model
get_3rd_layer_output = K.function([dnn_model.layers[0].input],
                                  [dnn_model.layers[3].output])

layer_output = {}
for ik in events:
    layer_output[ik] = get_3rd_layer_output([inputs[ik]])[0]
print layer_output['VBF'][0]
```

```python
neuron = 0

fig = pyp.figure()
fig.set_figheight(5)
fig.set_figwidth(7)

bins = np.linspace(0, 50, 100)
ic = 0
colors = ['black','blue','violet','green','orange','cyan','pink','yellow','red','gray','brown']
for ik in events:
    x = []
    for i in range(len(events[ik])):
        x.append(layer_output[ik][i][neuron])
    pyp.hist(x, bins, alpha=0.5, color=colors[ic], label=ik)
    ic += 1

pyp.xlabel('3rd layer output')
#pyp.ylabel('Normalized')
#pyp.ylim([0.001,10])
pyp.legend(loc='upper center')
pyp.gca().set_yscale("log")


pyp.tight_layout()
fig = pyp.show()
```

```python
spt = [[] for i in range(nparticles)]
seta = [[] for i in range(nparticles)]
sphi = [[] for i in range(nparticles)]
sw = [[] for i in range(nparticles)]

bpt = [[] for i in range(nparticles)]
beta = [[] for i in range(nparticles)]
bphi = [[] for i in range(nparticles)]
bw = [[] for i in range(nparticles)]

for ik in inputs:
    for i in range(len(inputs[ik])):
        for part in range(nparticles):
            pt  = inputs[ik][i][part*nfeatures]
            eta = inputs[ik][i][part*nfeatures+1]
            phi = inputs[ik][i][part*nfeatures+2]
            if(ik == 'VBF'):
                spt[part].append(pt)
                seta[part].append(eta)
                sphi[part].append(phi)
                sw[part].append(layer_output[ik][i][neuron])
            else:
                bpt[part].append(pt)
                beta[part].append(eta)
                bphi[part].append(phi)
                bw[part].append(layer_output[ik][i][neuron])

ymin = -0.05
ymax =  1.05
fig = pyp.figure()
fig.set_figheight(3)
fig.set_figwidth(17)                    
for i in range(7):
    fig.add_subplot(170+i+1)
    pyp.plot(bpt[i], bw[i], 'b.', ms=0.05)
    pyp.plot(spt[i], sw[i], 'r.', ms=0.05)
    pyp.xlabel('pT{0}'.format(i+1))
    pyp.xlim([-5,200])
    pyp.ylim([ymin,ymax])
pyp.tight_layout()
fig = pyp.show()

fig2 = pyp.figure()
fig2.set_figheight(3)
fig2.set_figwidth(17)                    
for i in range(7):
    fig2.add_subplot(170+i+1)
    pyp.plot(beta[i], bw[i], 'b.', ms=0.05)
    pyp.plot(seta[i], sw[i], 'r.', ms=0.05)
    pyp.xlabel('eta{0}'.format(i+1))
    #pyp.xlim([6,-6])
    pyp.ylim([ymin,ymax])
pyp.tight_layout()
fig2 = pyp.show()

fig3 = pyp.figure()
fig3.set_figheight(3)
fig3.set_figwidth(17)                    
for i in range(7):
    fig3.add_subplot(170+i+1)
    pyp.plot(bphi[i], bw[i], 'b.', ms=0.05)
    pyp.plot(sphi[i], sw[i], 'r.', ms=0.05)
    pyp.xlabel('phi{0}'.format(i+1))
    pyp.xlim([-4,4])
    pyp.ylim([ymin,ymax])
pyp.tight_layout()
fig3 = pyp.show()
```
