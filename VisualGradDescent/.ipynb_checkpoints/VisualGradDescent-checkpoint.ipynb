```python
#produces the distribution to fit
a = 0.5
b = 0.5

rd = ROOT.TRandom()
x = []
y = []
e = []
for i in range(100):
    e.append(0)
    ip = 1 - 1*rd.Rndm()
    x.append( ip )
    y.append( a+b*ip )
```

```python
#plot Loss space
lp1 = []
lp2 = []
lL = []

resolution = 200
lmin = 100000
p1min = 0
p2min = 0
for p1 in range(0,1):
    for r1 in range(resolution):
        ip1 = p1 + r1/float(resolution)
        
        gp1 = []
        gp2 = []
        gL = []
        for p2 in range(0,1):
            for r2 in range(resolution):
                ip2 = p2 + r2/float(resolution)
                for i in range(len(x)):
                    y = (a+b*x[i])
                    yl = max(0,(ip1+ip2*x[i]))
                    #mse
                    il = 0.5*pow(y-yl,2)
                    #mean_absolute_percentage_error
                    #il = math.fabs(y-yl)/float(y+0.00001)
                    #mine
                    #il = math.fabs(y-yl)
                    #il = math.sqrt(math.fabs(y-yl)/(y+yl))
                    #il = math.fabs((y-yl)/(y+yl))
                    
                    lp1.append( ip1 )
                    lp2.append( ip2 )
                    lL.append( il )
                    if(il < lmin):
                        lmin = il
                        p1min = ip1
                        p2min = ip2
                    #print "p1: %.3f, p2: %.3f, L: %.3f" % (ip1,ip2,il)
                    
        #lp1.append(gp1)
        #lp2.append(gp2)
        #lL.append(gL)
        
print "p1min: %.2f, p2min: %.2f, Lmin: %.2f" % (p1min, p2min, lmin)
```

```python
import theano
import theano.tensor as T

def myloss(y_true, y_pred):
    return T.sqrt(T.abs_((y_true-y_pred)/(y_true+y_pred)))
```
